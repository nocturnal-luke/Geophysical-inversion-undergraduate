# **Computing the Null Space from Row Echelon Form**

Given an \( m \times n \) matrix \( A \), the null space consists of all solutions to:

$$
A\mathbf{x} = 0
$$

## **Steps to Compute the Null Space**

1. **Convert \( A \) to Row Echelon Form (REF)**  
   - Apply Gaussian elimination to transform \( A \) into an upper triangular form.

2. **Identify Pivot and Free Variables**  
   - **Pivot variables** correspond to the leading (first nonzero) entries in each row.  
   - **Free variables** are the remaining columns (which do not contain a pivot).

3. **Express Pivot Variables in Terms of Free Variables**  
   - Solve for the pivot variables in terms of the free variables.
   - Assign free variables as parameters (let \( t, s, r, \dots \) be arbitrary values).

4. **Write the Null Space Basis Vectors**  
   - Each free variable generates a **basis vector** for the null space.
   - The null space is the span of these vectors.

---

## **Example Calculation**
Find the null space of:

$$
A = \begin{bmatrix} 1 & 2 & 3 & 4 \\ 0 & 1 & -1 & 2 \\ 0 & 0 & 0 & 0 \end{bmatrix}
$$

### **Step 1: Row Echelon Form (REF)**
The matrix is already in REF:

$$
\begin{bmatrix} 1 & 2 & 3 & 4 \\ 0 & 1 & -1 & 2 \\ 0 & 0 & 0 & 0 \end{bmatrix}
$$

### **Step 2: Identify Pivot and Free Variables**
- Pivot columns: **1st and 2nd** (leading ones)
- Free variables: **3rd and 4th** (\( x_3, x_4 \))

### **Step 3: Express Pivot Variables in Terms of Free Variables**
From row 1:

$$
x_1 + 2x_2 + 3x_3 + 4x_4 = 0 \quad \Rightarrow \quad x_1 = -2x_2 - 3x_3 - 4x_4
$$

From row 2:

$$
x_2 - x_3 + 2x_4 = 0 \quad \Rightarrow \quad x_2 = x_3 - 2x_4
$$

### **Step 4: Write Solution as a Linear Combination**
Let $x_3 = t$, $x_4 = s$. Substitute back:

$$
x_1 = -2(x_3 - 2x_4) - 3x_3 - 4x_4 = -5x_3 - 4x_4
$$

$$
x_2 = x_3 - 2x_4
$$

Solution as a vector:

$$
\begin{bmatrix} x_1 \\ x_2 \\ x_3 \\ x_4 \end{bmatrix}
=
x_3 \begin{bmatrix} -5 \\ 1 \\ 1 \\ 0 \end{bmatrix}
+
x_4 \begin{bmatrix} -4 \\ -2 \\ 0 \\ 1 \end{bmatrix}
$$

### **Final Null Space**
$$
\text{Null}(A) = \text{span} \left\{ \begin{bmatrix} -5 \\ 1 \\ 1 \\ 0 \end{bmatrix}, \begin{bmatrix} -4 \\ -2 \\ 0 \\ 1 \end{bmatrix} \right\}
$$

Since there are **2 free variables**, the null space has **dimension 2**, meaning it forms a **plane in $(\mathbb{R}^4)$**.
